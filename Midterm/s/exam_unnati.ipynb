```python
############################################################
# Exam.py 
# 
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  Alg
###########################################################    
class Alg():
    def __init__(self,a:'python list',ans:'python list',maxv:'list of size 1',work:'list of size 1',show:'bool'):
        self._a = a
        self._ans = ans
        self._maxv = maxv
        self._work = work
        self._show = show
        ## You can have your data structure here
        
        ## Nothing can be changed below
        self._exam() #Everything happens in _exam
        check_result(self._a,self._ans,self._maxv[0]) #your answer is checked here

    ############################################################
    #          Nothing can be changed in _exam
    ########################################################### 
    def _exam(self):
        self._alg()
        check_result(self._a,self._ans,self._maxv[0]) #your answer is checked here

    ############################################################
    #          WRITE CODE BELOW
    #  Populate self._maxv and self._work 
    #  if self._show is True, show steps of the algorithm
    ########################################################### 
    def _alg(self):
        #we have the array now
        #[] this defines-- 
        not_pick_arr = [] #-- Max value for this index if i not picked, was the last index picked or not
        pick_arr = [] #-- Max value for this index if i is picked, the last index will never be picked then this will always be zero

        if(self._show == True):
            print("Define two arrays , not_pick_arr and pick_arr for every element")
            print("Both the arrays have a two value array at every index [Max Value], [0 or 1 based on whether element before this was used to find Max or not")

        
        not_pick_arr.append([0,0])
        pick_arr.append([self._a[0],0])
        
        for i in range(1, len(self._a)):
            
            #increment work
            self._work[0]  =  self._work[0]+1 

            
           
            
            #calculating for not picked array for this i
            #when we do not pick, we can either take last or do not take even last - whatever is max and that defines my second value
            if(not_pick_arr[i-1][0] >= pick_arr[i-1][0]):
                #we do not pick the last
                if(self._show == True):
                    print("For element ", self._a[i], " populating not_pick_arr. We used the element before this to find Max ")
                    
                 
                max_val = not_pick_arr[i-1][0] 
                last_not_picked = 0
                not_pick_arr.append([max_val, last_not_picked])

            else:

                if(self._show == True):
                    print("For element", self._a[i], " populating not_pick_arr. We did use the element before this to find Max")
                    
                 
                max_val = pick_arr[i-1][0]
                last_was_picked = 1
                not_pick_arr.append([max_val, last_was_picked])
            
            

            #calculating max value if this is picked for i

            if(self._show == True):
                    print("For element",self._a[i], "Populating pick_arr. We can never use the element before this as if we pick i , we cannot pick i-1")

            
            val = self._a[i] + not_pick_arr[i-1][0] #max_value_when_last_was_not_picked
            last_picked = 0 #always
            pick_arr.append([val,0]) 


        if(self._show == True):
             print("Based on the last element not_pick_arr and pick_arr, Calculating the final answers")
                   
        # for j in not_pick_arr:
        #     print(j)

        # for j in pick_arr:
        #     print(j)
        #now we have calculated everything
        #Finding the max after the execution is done
        n = len(self._a)
        if(not_pick_arr[n-1][0] >= pick_arr[n-1][0]):
             #setting the maxValueOverall by not picking the last element in the array

            if(self._show == True):
                print("Final answer is calculated and last element is not included in the final answer")
             
            
            self._maxv[0] = not_pick_arr[n-1][0]
           
            compare_value = not_pick_arr[n-1][1]
            
            #finding the indices based on this when last element in the array was not picked
            for i in range(n-2,-1,-1):

                self._work[0]  =  self._work[0]+1  #increment work for calculating indices 
                
                if compare_value == 0 :
                    #last element was not picked
                    compare_value = not_pick_arr[i][1] #for next comparison
                else:
                    #last element was picked
                    self._ans.append(i)
                    compare_value = pick_arr[i][1]
            
        else:
            #max value is coming by picking the last element in the array
            #finding the indices

            if(self._show == True):
                print("Final answer is calculated and last element is included in the final answer")
            
            self._maxv[0] = pick_arr[n-1][0]
            
          
            self._ans.append(n-1)  #last element 
            #finding the indices based on this when last element in the array was Picked
            compare_value = pick_arr[n-1][1] #this can be zero or one, here always zero
            
            for i in range(n-2,-1,-1):

                self._work[0]  =  self._work[0]+1 #increment work for calculating indices
                
                if compare_value == 0 :
                    #last element was not picked
                    compare_value = not_pick_arr[i][1] #for next comparison
                else:
                    #last element was picked
                    self._ans.append(i)
                    compare_value = pick_arr[i][1]
                
        
        
            
        

############################################################
#  NOTHING CAN BE CHANGED BELOW
###########################################################  

############################################################
#  class  Solution
# Nothing can be changed in Solution
###########################################################    
class Solution():
    def XXX(self, nums:'Python list') -> 'int':
        #Nothing can be changedhere
        ans = []
        maxv = [0]
        work = [0]
        t = Alg(nums,ans,maxv,work,False)
        check_result(nums,ans,maxv[0])
        return maxv[0]

############################################################
# Nothing can be changed in check_result
# Note check_result is a global hanging function
###########################################################  
def check_result(a:'Python list',ans:'Python List',amax:'int'):
    x = sorted(ans)
    t = 0
    for e in x:
        t = t + a[e]
    assert(t == amax)
    # assert you did not break the rule
    lx = len(x)
    for i in range(0,lx-1,2):
        pos1 = x[i]
        pos2 = x[i +1]
        assert(pos2 >= (pos1+1))
 
```
